```python
# print the ImageNet class and it's wordnet description
from nltk.corpus import wordnet as wn 
from pytorch_pretrained_biggan.utils import IMAGENET 
synsets_col = []
for offset, class_id in IMAGENET.items():
    # print(offset, class_id)
    synsets = wn.synset_from_pos_and_offset('n',offset)
    synsets_col.append(synsets)
    # Get word descriptions for the synset ID
    print(f"ImageNet class ID {class_id}")
    print(synsets) # synsets.name()
    print("Definition:", synsets.definition())
```

```python
def search_hyponyms_for_synset(hypernym, synsets_col, show=False, prefix=""):
    if type(hypernym) == str:
        hypernym = wn.synset(hypernym)
    subclass = []
    for synsets in synsets_col:
        if hypernym in synsets.lowest_common_hypernyms(hypernym):
            subclass.append(synsets)
    if len(subclass) > 0:
        print(prefix, hypernym, f"has {len(subclass)} subclasses in the collection")
    if show:
        print(subclass)
    return subclass
```

```python
primate_subclass = search_hyponyms_for_synset('primate.n.02', synsets_col)
dog_subclass = search_hyponyms_for_synset('dog.n.01', synsets_col)
cat_subclass = search_hyponyms_for_synset('cat.n.01', synsets_col)
fish_subclass = search_hyponyms_for_synset('fish.n.01', synsets_col)
animal_subclass = search_hyponyms_for_synset('animal.n.01', synsets_col)
```

```python
def find_subclasses_recursive(node, synsets_col, prefix="", Ncutoff=1):
    subclasses = search_hyponyms_for_synset(node, synsets_col, prefix=prefix)
    if len(subclasses) > Ncutoff:
        for subclass in node.hyponyms():
            find_subclasses_recursive(subclass, subclasses, prefix=prefix+"  ", Ncutoff=Ncutoff)

root = wn.synset('entity.n.01')
node = root     
find_subclasses_recursive(root, synsets_col, Ncutoff=5)
```

```python
def find_subclasses_recursive(node, synsets_col, prefix=""):
    subclasses = search_hyponyms_for_synset(node, synsets_col, prefix=prefix)
    if len(subclasses) > 1:
        for subclass in node.hyponyms():
            find_subclasses_recursive(subclass, subclasses, prefix=prefix+"  ")

root = wn.synset('entity.n.01')
node = root     
find_subclasses_recursive(root, synsets_col)
```
